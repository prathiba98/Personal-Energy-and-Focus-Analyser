```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Date range for 1 year
dates = pd.date_range(start="2025-01-01", periods=365, freq="D")

data = []

for date in dates:
    day = date.day_name()
    is_workday = "Yes" if day not in ["Saturday", "Sunday"] else "No"

    sleep_hours = np.random.normal(7, 1.2)
    sleep_hours = max(4, min(sleep_hours, 9))  # realistic bounds

    mood_score = np.random.randint(2, 6)

    screen_time = np.random.normal(6, 1.5)
    screen_time = max(2, min(screen_time, 10))

    # Focus & energy logic
    energy_level = round(
        min(5, max(1, (sleep_hours / 2) - (screen_time / 4) + np.random.normal(1, 0.5)))
    )

    focus_level = round(
        min(5, max(1, energy_level + np.random.normal(0, 0.7)))
    )

    data.append([
        date,
        day,
        sleep_hours,
        mood_score,
        energy_level,
        focus_level,
        round(screen_time, 1),
        is_workday
    ])

columns = [
    "date",
    "day_of_week",
    "sleep_hours",
    "mood_score",
    "energy_level",
    "focus_level",
    "screen_time_hours",
    "workday"
]

df = pd.DataFrame(data, columns=columns)

# Save CSV
file_path = "personal_energy_focus_data.csv"
df.to_csv(file_path, index=False)

df.head(), file_path
```
